```python
# import libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data

df = pd.read_csv(r"C:\Users\DELL\Downloads\expenses.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# total expence calculation

total_expense = df['Amount'].sum()
```

```python
print(total_expense)
```

```python
# category wise expense

category_expense = df.groupby('Category')['Amount'].sum()
```

```python
category_expense
```

```python
# pie chart visualization

category_expense.plot(kind='pie', autopct="%1.1f%%")
plt.title("Expense Distribution by Category")
plt.show()
```

```python
# daily expense analysis

daily_expense = df.groupby("Date")["Amount"].sum()
daily_expense
```

```python
# bar chart

daily_expense.plot(kind="bar")
plt.title("Daily Expenses")
plt.xlabel("Date")
plt.ylabel("Amount")
plt.xticks(rotation=45)
plt.show()
```
